## Data Cleaning

### Case foling

```python
import string
import re
```

```python
def preprocess_text(kalimat):
    #mengubah kedalam huruf kecil semua
    lower_case=kalimat.lower()

    #menghapus angka
    hasil = re.sub(r"\d+", "", lower_case)

    #menghapus tanda baca
    hasil = re.sub(r'[\W_]+', ' ', kalimat)

    #menghapus spasi awal dan akhir
    hasil = hasil.strip()

    return hasil
```

```python
data['casefolding'] = data['content'].apply(preprocess_text)
data.head()
```

###  tokenizing Using NLTK

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
```

```python
def tokenize_text(kalimat):
    tokens = nltk.tokenize.word_tokenize(kalimat)
    return tokens
```

```python
data['tokenize'] = data['casefolding'].apply(tokenize_text)
data.head()
```

### Stopword removal (filtering)

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words = stopwords.words('indonesian')

def remove_stopwords(text):
    return [word for word in text if word not in stop_words]
```

```python
data['filtering'] = data['tokenize'].apply(lambda x: remove_stopwords(x))
```

```python
data.head()
```

### stemming

```python
factory = StemmerFactory()
stemmer = factory.create_stemmer()

def stem_text(text):
    return [stemmer.stem(word) for word in text]
```

```python
data['stemming_data'] = data['filtering'].apply(lambda x: ' '.join(stem_text(x)))
```
